```python
# Which videos were watched the longest
query2 = """

SELECT views.user_ID, views.view_date, views.duration_watched, views.feedback, videos.title, videos.category, videos.duration, videos.rating
FROM views
LEFT JOIN videos
ON views.video_ID = videos.pK
WHERE duration_watched > 15 
ORDER BY views.duration_watched DESC
"""
```

```python
result_df_2 = pd.read_sql(query2, con=connection)
result_df_2
```

```python
#average amount of subscribers grouped by subscribed and unsubscribed channels
query3 = """

SELECT subscribers.status, AVG(channels.subscribers) AS average_subscribers
FROM subscribers
LEFT JOIN channels
ON subscribers.channel_ID = channels.pK
GROUP BY subscribers.status
"""
```

```python
result_df_3 = pd.read_sql(query3, con=connection)
result_df_3
```

```python
#top five most popular videos by rating
query4 = """

SELECT videos.title, videos.rating, channels.name
FROM videos
LEFT JOIN channels
ON videos.channel_ID = channels.pK
ORDER BY rating Desc
LIMIT 5
"""
```

```python
result_df_4 = pd.read_sql(query4, con=connection)
result_df_4
```

```python
#Most popular channels by number of subscribers
query5 = """

SELECT name, subscribers
FROM channels
ORDER BY subscribers DESC
LIMIT 5
"""
```

```python
result_df_5 = pd.read_sql(query5, con=connection)
result_df_5
```

```python
#what video has the most views from the users we collected data on (top 5)
query6 = """

SELECT videos.title, COUNT(videos.title) as num_views
FROM views
LEFT JOIN videos
ON views.video_ID = videos.pK
GROUP BY videos.title
ORDER BY num_views Desc
LIMIT 5 
"""
```

```python
result_df_6 = pd.read_sql(query6, con=connection)
result_df_6
```
